```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df=pd.read_csv('Leo.csv')
df.index = np.arange(1, len(df) + 1)
```

```python
#Goal per season
ss1goal = len(df.loc[(df.Season == ('04/05'))])
ss2goal = len(df.loc[(df.Season == ('05/06'))])
ss3goal = len(df.loc[(df.Season == ('06/07'))])
ss4goal = len(df.loc[(df.Season == ('07/08'))])
ss5goal = len(df.loc[(df.Season == ('08/09'))])
ss6goal = len(df.loc[(df.Season == ('10/11'))])
ss7goal = len(df.loc[(df.Season == ('11-Dec'))])
ss8goal = len(df.loc[(df.Season == ('Dec-13'))])
ss9goal = len(df.loc[(df.Season == ('13/14'))])
ss10goal = len(df.loc[(df.Season == ('14/15'))])
ss11goal = len(df.loc[(df.Season == ('15/16'))])
ss12goal = len(df.loc[(df.Season == ('16/17'))])
ss13goal = len(df.loc[(df.Season == ('17/18'))])
ss14goal = len(df.loc[(df.Season == ('18/19'))])
ss15goal = len(df.loc[(df.Season == ('19/20'))])
ss16goal = len(df.loc[(df.Season == ('20/21'))])
ss17goal = len(df.loc[(df.Season == ('21/22'))])
ss18goal = len(df.loc[(df.Season == ('22/23'))])

goal_total = (ss1goal,ss2goal,ss3goal,ss4goal,ss5goal,ss6goal,ss7goal,ss8goal,ss9goal,ss10goal,ss11goal,ss12goal,ss13goal,ss14goal,ss15goal,ss16goal,ss17goal,ss18goal)
```

```python
#Goal per Season graph
x = np.array(goal_total)
y = np.array(['04/05','05/06','06/07','07/08','08/09','09/10','11/12','12/13','13/14','14/15','15/16','16/17','17/18','18/19','19/20','20/21','21/22','22/23'])
plt.title('All Lionel Messi goal')
plt.xlabel("Goal")
plt.ylabel("Seasons")
ind = np.arange(len(y))
plt.bar(ind,y,x,  color='purple')

plt.show()
```
