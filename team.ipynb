```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df=pd.read_csv('Leo.csv')
df.index = np.arange(1, len(df) + 1)
```

```python
barca = len(df.loc[(df.Club == ('FC Barcelona'))])
paris = len(df.loc[(df.Club == ('Paris Saint-Germain'))])
```

```python
x=np.array([barca,paris])
y=['barca','paris']
plt.pie(x,labels=y,shadow = True,autopct='%1.1f%%',startangle=90)
plt.axis('equal')
plt.title('Lionel Messi Team goal')
plt.show()
```
