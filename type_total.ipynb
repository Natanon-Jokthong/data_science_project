```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df=pd.read_csv('Leo.csv')
df.index = np.arange(1, len(df) + 1)
```

```python
lfs = len(df.loc[(df.Type == ('Left-footed shot'))])
rfs = len(df.loc[(df.Type == ('Right-footed shot'))])
penalty = len(df.loc[(df.Type == ('Penalty'))])
dfk = len(df.loc[(df.Type == ('Direct free kick'))])
head = len(df.loc[(df.Type == ('Header'))])
tap = len(df.loc[(df.Type == ('Tap-in'))])
chest = len(df.loc[(df.Type == ('Chest'))])

type_total = (lfs,rfs,penalty,dfk,head,tap,chest)
```

```python
x=np.array(type_total)
y=np.array(['Left foot shot','Right foot shot','penalty','Direct free kcik','Header','Tap','Chest'])
plt.barh(y,x)
plt.show()
```
